```python
import os
import shutil
import pandas as pd
```

```python


# Ruta a la carpeta local y destino
local_folder = "./SoundAnimals_Train/train_redimensionadas_subidas"

destination_folder = "./SoundAnimals_Train/train_redimensionadas_ultimo"

# Leer el listado de archivos faltantes
missing_files_csv = "./files_final.csv"
missing_files = pd.read_csv(missing_files_csv)['filename'].tolist()

# Crear la carpeta de destino si no existe
os.makedirs(destination_folder, exist_ok=True)

# Mover los archivos faltantes
for file in missing_files:
    source = os.path.join(local_folder, file)
    destination = os.path.join(destination_folder, file)
    if os.path.exists(source):
        shutil.move(source, destination)

print(f"Se han movido los archivos faltantes a la carpeta: {destination_folder}")
```

```python
import os
import pandas as pd

# Ruta a la carpeta local
folder_path = "./SoundAnimals_Train/train_redimensionadas_subidas"

# Ruta al archivo CSV
csv_path =  "./files_final.csv"

# Leer los nombres de los archivos desde el CSV
csv_files = pd.read_csv(csv_path)['filename'].tolist()  # Cambia 'filename' por el nombre exacto de la columna que contiene los nombres de los archivos

# Listar los nombres de los archivos en la carpeta local
local_files = os.listdir(folder_path)

# Comparar: Archivos que est√°n en el CSV pero no en la carpeta local
missing_files = [file for file in csv_files if file not in local_files]

# Mostrar los resultados
if missing_files:
    print("Los siguientes archivos est√°n en el CSV pero no en la carpeta local:")
    for file in missing_files:
        print(file)
else:
    print("Todos los archivos del CSV est√°n en la carpeta local.")

# Guardar los archivos faltantes en un archivo CSV (opcional)
output_csv_path = './missing_files.csv'
pd.DataFrame(missing_files, columns=['missing_filename']).to_csv(output_csv_path, index=False)
print(f"Se ha guardado la lista de archivos faltantes en: {output_csv_path}")
```

```python
import os
import shutil

# Ruta de la carpeta de origen (donde est√°n los archivos actualmente)

source_folder = "./SoundAnimals_Train/train_redimensionadas_subidas"

# Ruta de la carpeta de destino (donde quieres mover los archivos)
destination_folder = "./SoundAnimals_Train/train_redimensionadas_ultimo"

# Crear la carpeta de destino si no existe
os.makedirs(destination_folder, exist_ok=True)

# Listar los archivos en la carpeta de origen
files = os.listdir(source_folder)

# Mover archivos de origen a destino
for file in files:
    source_path = os.path.join(source_folder, file)
    destination_path = os.path.join(destination_folder, file)
    
    # Verificar que sea un archivo antes de mover
    if os.path.isfile(source_path):
        shutil.move(source_path, destination_path)
        print(f'Movido: {file}')

print('Todos los archivos han sido movidos.')
```
